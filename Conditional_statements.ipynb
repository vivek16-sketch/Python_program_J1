# Conditional statements in python 
1. Python have 4 types of conditional statements
2. `if` statement
3. `elif` statement
4. `else` statement
5. `nested` statements

```python if
if (condition):
    statements
    statements..
```

```python if,else
if (condition):
    statements
    statements...
else:
    statements
    statements...
```

```python if,elif,else
if (condition):
    statement
elif (condition):
    statements
elif (condition):
    statement
else:
    statements..
```

````python
#write a program to find the given number is positive or negitive or zero?
number = int(input("please enter number here:"))
if (number > 0):
    print(number,"is positive number")
elif (number < 0):
    print(number,"is negitive number")
else:
    print(number,"is zero")
````

````python
"""
write a program to find the student grades based on marks
1. if student marks in between 100 to 80 then grade is A
2. if student marks in between 79 to 60 then grade is B
3. if student marks in between 59 to 40 then grade is C
4. if student marks in between 39 to 25 then grade is D
5. if student marks in between 24 to 0 thenn grade is Fail
6. if other than above marks expected output should be please enter correct marks!
"""
marks=int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"is A grade")
elif (marks>=60 and marks<=79):
    print(marks,"is B grade")
elif (marks>=40 and marks<=59):
    print(marks,"is C grade")
elif (marks>=25 and marks<=39):
    print(marks,"is D grade") 
elif (marks>=0 and marks<=24):
    print(marks,"is Fail")
else:
    print(marks,"is not correct marks, please enter the correct marks")
````
